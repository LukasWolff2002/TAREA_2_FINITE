```python
import gmsh
import os

# Inicializar Gmsh
gmsh.initialize()
gmsh.model.add("modelo_llaves")

# Ruta del archivo .geo
ruta_geo = "malla.geo"

# Verifica si el archivo existe
if not os.path.isfile(ruta_geo):
    raise FileNotFoundError(f"âŒ No se encontrÃ³ el archivo: {ruta_geo}")

# Cargar geometrÃ­a desde archivo .geo
gmsh.merge(ruta_geo)

# ðŸ” Importante: sincronizar antes de mallar
gmsh.model.mesh.generate(3)  # â¬…ï¸ CAMBIO: generar malla 3D

# Exportar malla
ruta_salida = "malla.msh"
gmsh.write(ruta_salida)

print(f"âœ… Malla 3D generada y guardada como: {ruta_salida}")

# Finalizar Gmsh
gmsh.finalize()
```
