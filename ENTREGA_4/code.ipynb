```python
import gmsh
import os

# Inicializar Gmsh
gmsh.initialize()
gmsh.model.add("modelo_llaves")

# Ruta del archivo .geo
ruta_geo = "malla.geo"

# Verifica si el archivo existe
if not os.path.isfile(ruta_geo):
    raise FileNotFoundError(f"❌ No se encontró el archivo: {ruta_geo}")

# Cargar geometría desde archivo .geo
gmsh.merge(ruta_geo)

# 🔁 Importante: sincronizar antes de mallar
gmsh.model.mesh.generate(3)  # ⬅️ CAMBIO: generar malla 3D

# Exportar malla
ruta_salida = "malla.msh"
gmsh.write(ruta_salida)

print(f"✅ Malla 3D generada y guardada como: {ruta_salida}")

# Finalizar Gmsh
gmsh.finalize()
```
